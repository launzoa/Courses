```python
from selenium import webdriver
from selenium.webdriver.common.by import By


browser = webdriver.Edge()
browser.get('https://techstepacademy.com/trial-of-the-stones')


#Riddle of Stone // Objetivo: encontrar a senha secreta.
#fiz uma abreviação de cs = CSS_SELECTOR. Grande parte aqui é usada de css selector. 

input1_cs_element = browser.find_element(By.CSS_SELECTOR, "input#r1Input")
input1_cs_element.send_keys("rock")

button1_cs_element = browser.find_element(By.CSS_SELECTOR, "button#r1Btn")
button1_cs_element.click()

password_cs_element = browser.find_element(By.CSS_SELECTOR, "div#passwordBanner > h4")


#Riddle of Secrets // Objetivo: Utilizar a senha encontrada e validar essa parte do teste

input2_cs_element = browser.find_element(By.CSS_SELECTOR, "input#r2Input")
input2_cs_element.send_keys(password_cs_element.text)

button2_cs_element = browser.find_element(By.CSS_SELECTOR, "button#r2Butn")
button2_cs_element.click()


#The Two Merchants // Objetivo: Dados dois mercantes, encontrar qual deles tem mais dinheiro.
#xp é uma abreviação de XPATH

wealth_xp_element = browser.find_elements(By.XPATH, "//div")
merchant1 = wealth_xp_element[34]
merchant2 = wealth_xp_element[35]

wealth_merchant1 = merchant1.find_element(By.XPATH, ".//p")
wealth_merchant2 = merchant2.find_element(By.XPATH, ".//p")


if(wealth_merchant1.text > wealth_merchant2.text):
    richest_merchant = merchant1.find_element(By.XPATH, ".//span")
else:
    richest_merchant = merchant2.find_element(By.XPATH, ".//span")
    
result = browser.find_element(By.CSS_SELECTOR, "input#r3Input")
result.send_keys(richest_merchant.text)

button3_cs_element = browser.find_element(By.CSS_SELECTOR, "button#r3Butn")
button3_cs_element.click()


#Fim do desafio, botão para validar todas as respostas

button4_cs_element = browser.find_element(By.CSS_SELECTOR, "button#checkButn")
button4_cs_element.click()

complete_msg = browser.find_element(By.CSS_SELECTOR, "div#trialCompleteBanner > h4")
assert complete_msg.text == 'Trial Complete'

browser.quit()
```
