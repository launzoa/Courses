```python
'''
Esse exercício é sobre algumas propriedades comparativas entre objetos. Exemplo:
p1 = Product('milk', 5)
p2 = Product('bread', 11)
(p1 > p2) or (p1 < p2) etc...
'''

class Condicionais:
    
    
    def __init__(self, curso, nota):
        self.grade = []
        self.grade.append((curso, nota))
        self.__average__ = self.calculate_average()


    def __str__(self):

        return f"{self.grade[0:]}"


    def __call__(self, curso, nota):
        
        self.grade.append((curso, nota))
        self.__average__ = self.calculate_average()


    def calculate_average(self):
        avr = 0
        for block in self.grade:
            avr += block[1]
            
        avr /= len(self.grade)

        return avr
        
    @property
    def average(self):
     
        return self.__average__
    

    @average.setter
    def average(self, value):

        self.__average__ = value


    def __gt__(self, other):
        return self.__average__ > other.__average__
        

    def __lt__(self, other):
        return self.__average__ < other.__average__


    def __eq__(self, other):
        return self.__average__ == other.__average__


    def __ge__(self, other):
        return self.__average__ >= other.__average__


    def __le__(self, other):
        return self.__average__ <= other.__average__


    def __ne__(self, other):
        return self.__average__ != other.__average__
    

Aluno1 = Condicionais('Algoritmos', 10)
Aluno1('Cálculo', 10)

Aluno2 = Condicionais('Algoritmos', 9)
Aluno2('Cálculo', 10)

print(Aluno1 > Aluno2)
print(Aluno1 < Aluno2)
print(Aluno1 >= Aluno2)
print(Aluno1 <= Aluno2)
print(Aluno1 == Aluno2)
print(Aluno1 != Aluno2)

```
