```python
class CaesarCipher:
    
    
    letters = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m", "n", "o", "p", "q", "r", "s", "t", "u", "v", "w", "x", "y", "z"]
    
    def __init__(self, size):
        
        self.size = size
        

    
    def __call__(self, text):
        
        return self.cipher(text, self.size)
        
        
    def cipher(self, text, size):
        
        new_text = ""

        for let in text:
            if let == " ":
                new_text += " "
                
            else:
                for i in range(0,len(CaesarCipher.letters)):
                    
                    if (let == CaesarCipher.letters[i]):
                        new_text += CaesarCipher.letters[i-size]
        
        return new_text
    
                
    def cipher_list(self):
        
        letters_cipher = []
        
        for i in range(0, len(CaesarCipher.letters)):
            
            letters_cipher.append(CaesarCipher.letters[i-self.size]) 
            
        return letters_cipher
    
        
    def print_mapping(self):
        
        for i in range(0, len(CaesarCipher.letters)):
            print(f"{CaesarCipher.letters[i]} -> {self.cipher_list()[i]}")
            
        return None
    
    
    def decrypt(self, text):
        
        
        return self.cipher(text, -self.size)
    
    
if __name__ == '__main__':
   
    cipher = CaesarCipher(3)

    cipher("hello world")

    cipher.print_mapping()

    print(cipher.decrypt("ebiil tloia"))
```
